```python
print("If you put")
print("\033[33m", end="") #yellow
print("nothing as the")
print("\033[35m", end="") #purple
print("end character")
print("\033[32m", end="") #green
print("then you don't")
print("\033[0m", end="") #default
print("get odd gaps")
```

```python
import os

for i in range(1, 101):
  print(i)
  os.system("clear")
```

```python
import os, time

for i in range(1, 101):
  print(i)
  time.sleep(0.5)
  os.system("cls")
```

```python
import os, time
print('\033[?25l', end="")
for i in range(1, 101):
  print(i)
  time.sleep(0.2)
  os.system("clear")
```

```python
def newtext(color, text):
  if color == "red":
    print("\033[31m", text, end="",sep="")
  elif color == "green":
    print("\033[32m", text,end="",sep="")
  elif color == "blue":
    print("\033[34m", text,end="",sep="")
  else:
    print("\033[0m", text,end="",sep="")


print("Super Subroutine")
print()
print("With my ", end="")
newtext("red", "new program ")
newtext("green", "I can just call ")
newtext("blue", "text.")
```
